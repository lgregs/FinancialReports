### Base de Ações

```python
acoes = ['GOLL4.SA', 'CVCB3.SA', 'WEGE3.SA', 'MGLU3.SA', 'TOTS3.SA', 'VALE3.SA', 'PETR3.SA', 'ABEV3.SA', 'ITSA3.SA', 'BOVA11.SA']
```

```python
acoes_df = pd.DataFrame()
for acao in acoes:
  acoes_df[acao] = yf.download(acao, start='2010-01-01')['Close']
```

```python
acoes_df
```

```python
acoes_df.isnull().sum()
```

```python
acoes_df.describe()
```

```python
plt.figure(figsize=(15,55))
i = 1
for i in np.arange(1, len(acoes_df.columns)):
  plt.subplot(11, 1, i + 1)
  sns.boxplot(x = acoes_df[acoes_df.columns[i]])
  plt.title(acoes_df.columns[i])
```

```python
acoes_df.plot(x = 'Date', figsize = (15,10), title = 'Histórico do preço das ações');
```

```python
acoes_normalized = acoes_df.copy()
for i in acoes_normalized.columns[1:]:
  acoes_normalized[i] = acoes_normalized[i]/acoes_normalized[i][0]
```

```python
acoes_normalized
```

```python
acoes_normalized.plot(x = 'Date', figsize = (15,10), title = 'Histórico do preço das ações - Normalizadas');
```
